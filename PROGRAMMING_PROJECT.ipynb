Duration

```python
#Average Duration of Top Hit Songs for each year:
#converting the duration of songs from milliseconds to minutes for easier understanding
def time_convert(ms):
    sec=ms/1000
    return f"{int(sec//60)}:{int(sec%60)}"
durations = df[['duration_ms','year']].groupby('year').mean().reset_index().iloc[0:20]
durations['min:sec'] = durations['duration_ms'].apply(time_convert)
durations
```

```python
# Creating a function that converts milliseconds to minutes and seconds
def ms_to_min_sec(ms):
    sec = ms/1000
    return f"{int(sec//60)}:{int(sec%60)}"

df['min:sec'] = df['duration_ms'].apply(ms_to_min_sec)
```

```python
#Longest songs
songs_by_duration = df.sort_values("min:sec", ascending=False)[["song", "artist", "min:sec"]]
songs_by_duration[:5]
```

```python
from kmodes.kmodes import KModes
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
#Elbow curve to find optimal K
cost = []
K = range(1,20)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init = 'random', n_init = 5, verbose=1)
    kmode.fit_predict(df)
    cost.append(kmode.cost_)
    
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Building the model with 5 clusters
kmode = KModes(n_clusters=4, init = "random", n_init = 5, verbose=1)
clusters = kmode.fit_predict(df)
clusters
```

```python
df.insert(0, 'Cluster', clusters, True)
df
```

```python
plt.figure(figsize=(20,16))
plt.scatter(df['energy'], df['loudness'])
plt.xlabel('Energy')
plt.ylabel('Loudness')
```

```python
square_distances = []
x = df[['energy','loudness']]
for i in range(1, 11):
    km = KMeans(n_clusters=5, random_state=42)
    km.fit(x)
    square_distances.append(km.inertia_)
```

```python
km = KMeans(n_clusters=5, random_state=42)
y_pred = km.fit_predict(x)
```

```python
plt.figure(figsize=(30,30))
labels = ['pop', 'hip-hop', 'R&B', 'deance', 'rock']
for i in range(5):
    plt.scatter(x.loc[y_pred==i, 'energy'], x.loc[y_pred==i, 'loudness'], label=labels[i])

plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.legend()
plt.show()
```
